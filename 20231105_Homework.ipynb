# 영상 처리하기

```python
try:
    cap=cv2.VideoCapture('1.mp4')
    print('동영상 읽기 성공')
except:
    print('동영상 읽기 실패')
```

```python
while True:
    ret, frame = cap.read()
    #동영상에서 이미지를 프레임 단위로 읽어옴
    #ret 잘 읽었는지 T/F로 표현
    #frame 1프레임의 이미지 저장
    
    if not ret : 
        #ret이 false일 때 (=파일에 문제가 있거나 동영상이 끝남)
        print('이미지 읽기 실패 또는 모두 읽음')
        cap.release()
        cv2.destroyAllWindows()
        #윈도우 실행창 종료
        break
        #반복문 종료
    
    cv2. imshow('animation', frame)
    #동영상을 잘 읽을 때
    
    k = cv2.waitKey(33)
    
    if k == 49: 
        # 49 -> ASCII코드로 1을 의미
        print('동영상 종료')
        cap.release()
        # 비디오 종료
        cv2.destroyAllWindows()
        # 윈도우 실행창 종료
        break
        # 1을 누르면 동영상 종료
    
    #cv2.waitKey(33)
    #한 장의 사진을 0.033초 동안 띄움
```

```python
#실패
#동영상 녹화하기

try:
    cap = cv2.VideoCapture('1,mp4')
    print('동영상 읽기 성공')
except:
    print('동영상 읽기 실패')
    

#동영상 녹화 설정
fps = 30.0
width = int(cap.get(3))
height = int(cap.get(4))

#코덱 설정 (동영상 압축 방법)
#참고 : mp4 >> MPEG, avi >> DIVX
fcc = cv2.VideoWriter_fourcc(*'MPGE')

#녹화 저장 설정
#cv2.VideoWriter(저장할 파일명, 코덱, 프레임 수, 동영상 크기)
out = cv2.VideoWriter('1_result.mp4', fcc, fps, (width, height))

record = False

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('이미지 읽기 실패 또는 모두 읽음')
        cap.release()
        cv2.destroyAllWindows()
        break
    
    cv2.imshow('animation', frame)
    out.write(frame)
    
    k = cv.waitKey(33)
    
    if k == ord(str(1)):
        print('동영상 종료')
        cap.release()
        cv2.destroyAllWindows()
        break
        
    if k == ord(str(2)):
        record = True
        
    if record:
        out.write(frame)
        
    if k == ord(str(3)):
        out.release()
        print('녹화 종료')
```

```python
#동영상 녹화 재시도 성공본

try:
    cap=cv2.VideoCapture('1.mp4')
    print('동영상 읽기 성공')
except:
    print('동영상 읽기 실패')

fps = 30.0
width = int(cap.get(3))
height = int(cap.get(4))
fcc = cv2.VideoWriter_fourcc(*'MPEG')
out = cv2.VideoWriter('1_result.mp4', fcc, fps, (width, height))

record = False

while True:
    ret, frame = cap.read()

    if not ret : 
        print('이미지 읽기 실패 또는 모두 읽음')
        cap.release()
        cv2.destroyAllWindows()
        break

    
    cv2. imshow('animation', frame)
    out.write(frame)
    
    k = cv2.waitKey(33)
    
    if k == 49: 
        print('동영상 종료')
        cap.release()
        cv2.destroyAllWindows()
        break
        
    if k == 50:
        record = True
        print('녹화 시작')
    
    if record : 
        out.write(frame)
        
    if k == 51:
        out.release()
        print('녹화 종료')
 

```

```python
#영상 캡쳐해서 이미지로 저장하기

#cv.imwrite(저장할 파일명, 저장할 이미지, 압축 설정)
#IMWRITE_JPEG_PROGRESIVE : jpg파일로 이미지 저장
#0 압축률 (0-9) : 0은 압축하지 않음을 의미

try:
    cap = cv2.VideoCapture('1.mp4')
    print('동영상 읽기 성공')
except:
    print('동영상 읽기 실패')

num = 0
#파일 이름 저장용
out = cv2.VideoWriter()

##이게 맞는건지는 모르겠지만 일단 해보자
fps = 30.0
width = int(cap.get(3))
height = int(cap.get(4))
fcc = cv2.VideoWriter_fourcc(*'MPEG')
out = cv2.VideoWriter('1_result1.mp4', fcc, fps, (width,height))

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('이미지 읽기 실패 또는 모두 읽음')
        cap.release()
        cv2.destroyAllWindows()
        break
        
    cv2.imshow('animation', frame)
    out.write(frame)
    
    k = cv2.waitKey(33)
    
    if k == 49:
        print('동영상 종료')
        cap.release()
        cv2.destroyAllWindows()
        break
    
    if k == 50:
        cv2.imwrite('./cap_img/cap_img{}.jpg'.format(num), frame,
                  params = [cv2.IMWRITE_JPEG_PROGRESSIVE, 0])
        print('./cap_img/cap_img{}.jpg 파일 저장'.format(num))
        num += 1
```
